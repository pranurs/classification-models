```python
def add_to_dict (dictionary, sentence):
    for word in sentence:
        if word not in dictionary:
            dictionary[word] = 1
        else:
            dictionary[word] += 1
    return dictionary
```

```python
positive_dict = {}
negative_dict = {}
overall_dict = {}
# num_positive = 0
# num_negative = 0

for i in range(len(T)):
    if T[i] is 1:
        positive_dict = add_to_dict(positive_dict, X[i])
    else:
        negative_dict = add_to_dict(negative_dict, X[i])
    overall_dict = add_to_dict(overall_dict, X[i])
```

```python
negative_dict
```

```python
P_pos = np.sum(T)/len(T)
P_neg = 1 - P_pos
```

```python
print(P_pos, P_neg)
```

```python
def make_class_prediction (sentence, dictionary, class_prob, class_len):
    
    prediction = 0
    
    for word in sentence:
        if word not in dictionary:
            count = 0
        else:
            count = dictionary[word]
            
        P = ((count) + 1) / (class_len + 2)
        
        prediction += np.log(P)
        
    prediction += np.log(class_prob)
    return prediction    
```

```python
pos = make_class_prediction(X[0], positive_dict, P_pos, np.sum(T))
```

```python
neg = make_class_prediction(X[0], negative_dict, P_neg, len(T) - np.sum(T))
```

```python
print(pos, neg)
```

```python
accuracy = 0
for i in range(len(T)):
    pos = make_class_prediction(X[i], positive_dict, P_pos, np.sum(T))    
    neg = make_class_prediction(X[i], negative_dict, P_neg, len(T) - np.sum(T))
    if(pos>neg):
        y = 1
    else:
        y = 0
    accuracy += (y == T[i])
accuracy /= len(T)
```

## Making vocabulary set

```python
vocabulary = set()
for i in range(len(X)):
    for word in X[i]:
        vocabulary.add(word)
```

```python
len(vocabulary)
```

## Bag of Words model for each class

```python
classes = set()
documents_in_class = {}
for i in range(len(T)):
    classes.add(T[i])
    if T[i] not in documents_in_class:
        documents_in_class[T[i]] = []
    documents_in_class[T[i]].append(X[i])
```

```python
len(documents_in_class[0])
```

```python
len(documents_in_class[1])
```

```python
classes
```

```python
bow = {}
word_total = {}
class_prob = {}
```

```python
for c in classes:
    
    if c not in bow:
        bow[c] = {}
    if c not in word_total:
        word_total[c] = 0
    if c not in class_prob:
        class_prob[c] = 0
        
    for doc in documents_in_class[c]:
        
        for word in doc:
            if word not in bow[c]:
                bow[c][word] = 0
                
            bow[c][word] += 1
            word_total[c] += 1
            
        class_prob[c] += 1
    
    class_prob[c] /= len(T)
```

## Calculating P(w|c) for each word in the vocabulary, for each class

```python
P_w_c = {}
for word in vocabulary:
    
    for c in classes:
        
        if word not in bow[c]:
            bow[c][word] = 0
            
        P_w_c[(word,c)] = (bow[c][word] + 1) / (word_total[c] + len(vocabulary))
```

## Testing the classifier and making predictions

```python
accuracy = 0
for i in range(len(T)):
    
    classify = {}
    argmax = -1e6
    prediction = -1
    
    unique_words = set(X[i])
#     print(X[i])
#     print(unique_words)
    
    for word in unique_words:
        
        if word not in vocabulary:
            continue
            
        for c in classes:
            
            if c not in classify:
                classify[c] = np.log(class_prob[c])
                
            classify[c] += np.log(P_w_c[(word,c)])
            
#             print(word, c, classify[c])
        
    for c in classes:
        
#         print(classify[c])
        if classify[c] > argmax:
            argmax = classify[c]
            prediction = c
            
    accuracy += (T[i] == prediction)

accuracy /= len(T)
```
